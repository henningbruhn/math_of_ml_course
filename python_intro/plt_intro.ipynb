# Short intro to matplotlib

matplotlib is the main package in python to plot data. (There are others, some of which are more sophisticated.) Often it's imported as <code>plt</code>.

```python
import matplotlib.pyplot as plt  ### This looks a bit more complicated than necessary because of bad design decisions
# "as" simply defines an alias, an alternative name
# and indeed, always typing matplotlib.pyplot.some_thing would be very cumbersome
# easier is plt.some_thing
```

To demonstrate how matplotlib works I will use another standard python package, namely numpy. 

```python
import numpy as np  # numpy is often imported as np
```

We need some data to plot. Let's first generate some x-values. For this we need the numpy method <code>linspace</code>, which simply returns a number of equidistant x-values between a lower bound and an upper bound.

```python
xx1=np.linspace(0,20,100)  ## generate 100 equidistant values between 0 and 20
xx1
```

By the way, what we see here is a numpy-array. These are better suited for scientific calculations than the standard lists. 

We next generate some y-values by feeding the x-values into a sine function.

```python
yy1=np.sin(xx1)
yy1
```

Next, we can actually plot. There are several different ways to do so. I will present here one that easily generalises to plots that have several different subplots. First we instantiate the plot with <code>plt.subplots()</code>. This returns two objects, a handle to the whole figure and a handle to the so-called *axis*. The axis <code>ax</code> is here the more interesting object, as we need it to plot. This we do with <code>ax.plot(x_values,y_values)</code>.

```python
fig,ax=plt.subplots()
ax.plot(xx1,yy1)
```

There are several options to customise the plot. First, we might want to control the size of the plot, which we can with the parameter <code>figsize=(10,5)</code> that we pass to <code>subplots</code>. Next, we'd like to change the style and colour of the line. We make it red and dashed by adding <code>'r--'</code> to the plot command. Other options would have been 'b-' (solid blue line), 'k:' (dotted in black) and 'go-' (green, solid line, data points as dots) -- try experimenting and also see the [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html). For good measure we also use a thicker line. 

```python
fig,ax=plt.subplots(figsize=(10,5))
ax.plot(xx1,yy1,'r--',linewidth=4)
```

What is missing now are labels. 

```python
fig,ax=plt.subplots(figsize=(10,5))
ax.plot(xx1,yy1,'r--',linewidth=4)
ax.set_xlabel("The x-values")
ax.set_ylabel("sin(x)")
ax.set_title("A sine function in red!")
```

There are many more options: You can change the colour of the back ground, the frame can be turned off, the size of the labels can be changed etc. Instead of going through all the options let's change most of them at once. <code>matplotlib</code> offers several styles that change the appearance of all plots. Most of the styles look quite a bit better than the default -- so let's use them.

```python
plt.style.use('seaborn')  ## seaborn is one of many available styles
fig,ax=plt.subplots(figsize=(10,5))
ax.plot(xx1,yy1,linewidth=4)
ax.set_xlabel("The x-values")
ax.set_ylabel("sin(x)")
ax.set_title("A sine function")
```

A list of available styles can be found [here](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html).

Let's add some more lines to the plot. 

```python
yy2=np.cos(xx1)
yy3=0.1*xx1-1

fig,ax=plt.subplots(figsize=(10,5))
ax.plot(xx1,yy1,linewidth=4)
ax.plot(xx1,yy2,linewidth=4)
ax.plot(xx1,yy3,linewidth=4)
ax.set_xlabel("The x-values")
ax.set_ylabel("sin(x)")
ax.set_title("A sine function")
```

Note how automatically the different lines get coloured differently. Note moreover that the seaborn style is still in use even though we did not specify it again.

Next, let's try more ways to plot data. We do a scatter plot.

```python
## let's generate some random data
n=500
xx4=np.random.random(n)
yy4=xx4+0.2*np.random.normal(size=n)

## next, we plot
fig,ax=plt.subplots()
ax.scatter(xx4,yy4)
```

Again, there are many options to control the appearance of the plot. Let's list the perhaps four most important ones.

- the used markers: [here](https://matplotlib.org/stable/api/markers_api.html) is a list, but common ones are 'x', '+', 'o' and '*'
- the size of the markers: controlled by the parameter <code>s</code>
- the colour of the markers: controlled by the parameter <code>c</code>
- opacity: controlled by the parameter <code>alpha</code>

```python
fig,ax=plt.subplots()
ax.scatter(xx4,yy4,c='k',s=100,alpha=0.4)
```

Often we have some distributed data such as the heights of individuals or house prices. In this cases it's useful to plot a histogram of the distribution.

```python
## let's generate some normally distributed data
yy5=np.random.normal(size=1000)

## next, we plot
fig,ax=plt.subplots()
ax.hist(yy5)
plt.show()  ## this is only to prevent some garbage output
```

The most useful parameter for <code>ax.hist</code> is the number of bins. To see where one bar ends and the next starts, set the edge colour of the bars with <code>edgecolor</code>.

```python
fig,ax=plt.subplots()
ax.hist(yy5,bins=50,edgecolor='w')  ## we set the edge colour to white, ie 'w'
plt.show()  ## this is only to prevent some garbage output
```

As a final type of plot let's do bar plots. (Note: nobody, and I cannot stress this enough, likes pie plots. They are an abomination.)

```python
## first let's generate some data
data=[12,42,27,29,2]
labels=['rats','cats','pythons','squirrels','dogs']

## next, we plot
fig,ax=plt.subplots()
ax.bar(labels,data)
ax.set_title('awesomeness of different pets')
```

Sometimes a horizontal bar plot is nicer.

```python
## next, we plot
fig,ax=plt.subplots()
ax.barh(labels,data)
ax.tick_params(axis='y', labelsize=15) ## let's have larger labels
ax.set_title('awesomeness of different pets',fontsize=15) ## let's also have a larger title
```

For our final trick, we look at how several plots can be combined to a single one. Indeed <code>plt.subplots</code> can be told to set up several subplots and in what kind of configuration. Here is how it's done: <code>plt.subplots(rows,columns)</code> sets up <code>rows</code>$\times$<code>columns</code> many plots in <code>rows</code> rows and <code>columns</code> columns. It returns a hangle <code>fig</code> to the whole figure and an array <code>axs</code> to the axis. In our next example we put all plots in one row. As a result <code>axs</code> is a one-dimensional array.

```python
fig,axs=plt.subplots(1,4,figsize=(20,5))
axs[0].plot(xx1,yy1,linewidth=4)
axs[0].plot(xx1,yy2,linewidth=4)
axs[0].plot(xx1,yy3,linewidth=4)

axs[1].scatter(xx4,yy4,c='k',s=100,alpha=0.4)

axs[2].hist(yy5,bins=50,edgecolor='w')  

axs[3].barh(labels,data)
```

Let's do this again, but this time let's do a 2x2 grid. Also let's save the figure at the end via <code>plt.savefig</code>. 

```python
fig,axs=plt.subplots(2,2,figsize=(10,10))
axs[0,0].plot(xx1,yy1,linewidth=4)  ## note that axs is now a 2-dim array
axs[0,0].plot(xx1,yy2,linewidth=4)
axs[0,0].plot(xx1,yy3,linewidth=4)

axs[0,1].scatter(xx4,yy4,c='k',s=100,alpha=0.4)

axs[1,0].hist(yy5,bins=50,edgecolor='w')  

axs[1,1].barh(labels,data)

## to save we need to specify the name of the file
## note that the extension, here .png, indicates the format
## dpi, ie, dots per inch, controls the resolution
plt.savefig("nice_plots.png",dpi=300)
```

This concludes the short introduction to matplotlib. For more, as always, have a look at the [documentation](https://matplotlib.org/stable/contents.html).
