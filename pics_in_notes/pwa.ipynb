# piecewise affine functions

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("lecture.mplstyle")
```

```python
def fun(x):
    if x<=2:
        return 1*x+0
    if x<=3:
        return -0.5*(x-2)+2
    return 0.5*(x-3)+1.5

def conc(x):
    if x<=2:
        return 1.5*x-1
    return 2

def conv(x):
    if x<=3:
        return -0.5*(x-2)+0
    return 0.5*(x-3)-0.5
xx=np.linspace(0,5,200)
yy=[fun(x) for x in xx]
y1=[conc(x) for x in xx]
y2=[conv(x) for x in xx]
```

```python
ann_fontsize=16
fig,axs=plt.subplots(1,2,figsize=(8,4),sharey=True)
axs[0].plot(xx,yy,'k',linewidth=3,label='$f$')
#axs[0].plot(xx,y2,'k-.',linewidth=3,label='$g$',alpha=0.5)
#axs[0].plot(xx,y1,'k--',linewidth=3,label='$h$',alpha=0.5)
axs[0].plot(xx,y2,'k-',linewidth=3,label='$g$',alpha=0.5)
axs[0].plot(xx,y1,'k-',linewidth=3,label='$h$',alpha=0.5)
axs[0].set_ylim(-1.1,3.1)
axs[0].set_xlim(-0.1,5.1)
axs[0].set_yticks(range(-1,4))
axs[0].set_xticks(range(0,6))
#axs[0].legend(handlelength=3,fontsize=16)
axs[0].annotate("$f$",(3.2,1.35),fontsize=ann_fontsize)
axs[0].annotate("$g$",(4.4,0.4),fontsize=ann_fontsize)
axs[0].annotate("$h$",(0.5,-0.55),fontsize=ann_fontsize)

axs[1].plot(xx,1.5*xx-1,'k--',linewidth=3,alpha=0.7)
axs[1].plot(xx,0*xx+2,'k--',linewidth=3,alpha=0.7)
axs[1].plot(xx,y1,'k',linewidth=3,label='g',alpha=0.5)
axs[1].plot(xx,-0.5*(xx-2)+0,'k--',linewidth=3,alpha=0.7)
axs[1].plot(xx,0.5*(xx-3)-0.5,'k--',linewidth=3,alpha=0.7)
axs[1].plot(xx,y2,'k',linewidth=3,label='h',alpha=0.5)
axs[1].set_ylim(-1.1,3.1)
axs[1].set_xlim(-0.1,5.1)
axs[1].set_yticks(range(-1,4))
axs[1].set_xticks(range(0,6))
axs[1].annotate("$h_1$",(0.7,2.1),fontsize=ann_fontsize)
axs[1].annotate("$h_2$",(2.5,2.4),fontsize=ann_fontsize)
axs[1].annotate("$g_1$",(2.1,-0.7),fontsize=ann_fontsize)
axs[1].annotate("$g_2$",(3.7,-0.7),fontsize=ann_fontsize)

plt.tight_layout()
plt.savefig("pw_aff.png",dpi=300)
```
