# Loss functions

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("lecture.mplstyle")
```

```python
_,ax=plt.subplots(figsize=(8,4))
xx=np.linspace(-2.1,2.1,200)
ax.plot(xx,xx<0,"k-",label="zero-one")
ax.plot(xx,np.exp(-xx),"k--",label="exponential")
ax.plot(xx,(1-xx)**2,"k-.",label="square")
ax.plot(xx,np.log2(1+np.exp(-xx)),"k:",label="cross-entropy")
ax.set_ylim(-0.5,4.5)
ax.legend(handlelength=3)
plt.savefig("loss_functions.png",dpi=300)
```

```python
_,ax=plt.subplots(figsize=(8,4))
xx=np.linspace(-2.1,2.1,200)
ax.plot(xx,xx<0,"k-",label="zero-one")
ax.plot(xx,np.log2(1+np.exp(-xx)),"k-",label="logistic",alpha=0.5)
ax.plot(xx,(1-xx)**2,"k--",label="square",alpha=0.5)
ax.plot(xx,np.exp(-xx),"k-.",label="exponential",alpha=0.5)
ax.set_ylim(-0.3,4.3)
ax.legend(handlelength=3,fontsize=14)
plt.savefig("loss_functions.png",dpi=300)
```

## Regression losses

```python
mse=lambda x:x**2
mae=lambda x: np.abs(x)
def huber(x):
    return np.where(np.abs(x)<1, 0.5*x**2,np.abs(x)-0.5)
```

```python
_,ax=plt.subplots(figsize=(8,4))
xx=np.linspace(-2.1,2.1,200)
ax.plot(xx,mse(xx),'k-',label="MSE")
ax.plot(xx,mae(xx),'k--',label="MAE")
ax.plot(xx,huber(xx),'k-.',label="Huber")
ax.legend(handlelength=3,fontsize=14)
plt.savefig("regression_losses.png",dpi=300)
```
