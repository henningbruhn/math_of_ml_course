# Height of men and women


We look at the distribution of heights of American men and women of 2007 / 2008. Sources are found here:

https://www2.census.gov/library/publications/2010/compendia/statab/130ed/tables/11s0205.pdf

https://www2.census.gov/library/publications/2010/compendia/statab/130ed/tables/

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")
```

```python
heights_raw=pd.read_csv("height_distribution.csv")
heights_raw
```

```python
heights_raw=heights_raw.drop(labels=[1,5,12,19])
heights_raw
```

```python
heights_raw=heights_raw.replace("(B)",0)
heights_raw=heights_raw.replace("-",0)
heights_raw
```

```python
def replacer(x):
    if not type(x) is str:
        return x
    else:
        return x.replace("\\1 ","")
heights_raw=heights_raw.applymap(replacer)
heights_raw
```

```python
def convert(feet,inches):
    return 0.3048*feet+25.4/1000*inches

import re
import numpy as np

def metricise(s):
    if not type(s) is str:
        return s
    feet_search=re.search(r"(\d) feet",s)
    if feet_search is None:
        return np.nan
    else:
        feet=int(feet_search.group(1))
    inches_search=re.search(r"(\d+) inch",s)
    if inches_search is None:
        inches=0
    else:
        inches=int(inches_search.group(1))
    return convert(feet,inches)    

metricise("blubb")
```

```python
heights_raw["Height"]=heights_raw[["Height"]].applymap(metricise)
heights_raw
```

```python
males_raw=heights_raw.iloc[:,:7]
females_raw=heights_raw.iloc[:,[0,7,8,9,10,11,12,13]]
females_raw
```

```python
def clean(df):
    columns=df.iloc[0].copy()
    columns[0]="Height"
    df.columns=columns
    df=df.rename(columns={np.nan:'Height'})
    df=df.drop(labels=[0])
    for col in df.columns:
        df[col]=df[col].astype(float)
    return df

women=clean(females_raw)
men=clean(males_raw)
women.info()
```

```python
plt.plot(women['Height'],women["40 to 49 years"])
plt.plot(men['Height'],men["40 to 49 years"])
```

```python
def de_cumsum(df):
    for col in df.columns:
        if col=="Height":
            continue
        l=list(df[col][:-1])
        l.insert(0,0)
        s=pd.Series(l,index=df.index)
        df[col]=df[col]-s

de_cumsum(women)
de_cumsum(men)
```

```python
plt.plot(women['Height'],women["40 to 49 years"])
plt.plot(men['Height'],men["40 to 49 years"])
```

```python
from scipy.stats import norm

def data_and_norm(df,col="40 to 49 years"):
    data=[[height]*round(prob*10) for height,prob in zip(df["Height"],df[col])]
    data=[e for l in data for e in l]
    # Fit a normal distribution to the data:
    mu, std = norm.fit(data)
    rv=norm(loc=mu,scale=std)
    return data,rv

men_data,men_fit=data_and_norm(men)
women_data,women_fit=data_and_norm(women)
```

```python
scaling_factor=max(women["Height"])-min(women["Height"])
bar_width=scaling_factor/len(women["Height"])
bar_width
```

```python
_,ax=plt.subplots(figsize=(5,3))

malecolour='b'
femalecolour='g'

ax.bar(women['Height'],women["40 to 49 years"],color=femalecolour,edgecolor='k',width=0.02,alpha=0.3)
ax.bar(men['Height'],men["40 to 49 years"],color=malecolour,edgecolor='k',width=0.02,alpha=0.)
ax.plot(x, women_fit.pdf(x)*bar_width*100, femalecolour, linewidth=4,label="women",alpha=0.9)
ax.plot(x, men_fit.pdf(x)*bar_width*100, malecolour, linewidth=4,label="men",alpha=0.9)

ax.legend(fontsize=12)
ax.set_xlabel("Height (m)",fontsize=12)
ax.set_ylabel("Percentage",fontsize=12)
plt.tight_layout()
plt.savefig("heights.png",dpi=300)
```
