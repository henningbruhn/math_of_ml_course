# More data -- sizes of facial recognition data sets by year

```python
### code inspired by https://pbpython.com/pandas-html-table.html
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("lecture.mplstyle")


```

```python
try:
    data=np.loadtxt("facedatasets.dat")
except:
    #tables = pd.read_html('https://en.wikipedia.org/wiki/List_of_datasets_for_machine-learning_research')
    tables = pd.read_html('https://en.wikipedia.org/wiki/List_of_datasets_in_computer_vision_and_image_processing')
    faces=tables[4]
    faces['num_inst'] = faces['Instances'].replace({'[^0-9]': '', '~': ''}, regex=True).astype('int')
    faces.loc[19,'num_inst']=45592
    data=faces[["Created (updated)","num_inst"]].to_numpy()
    np.savetxt("facedatasets.dat",data)

plt.figure(figsize=(3,3))
plt.scatter(data[:,0],data[:,1],color='k',zorder=4,edgecolor='w')
plt.ylabel("size of data set")
plt.yscale('log')
plt.title('facial recognition')
plt.savefig('datasetsize.png',dpi=300)
```
