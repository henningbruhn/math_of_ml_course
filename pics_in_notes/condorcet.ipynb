# Plots for bagging classifiers

```python
import numpy as np
import sklearn 
import matplotlib.pyplot as plt
import math

plt.style.use("lecture.mplstyle")
```

## correlated estimators

```python
def lower_bound_fun(c):
    return lambda p:(p-0.5)**2/(c*p*(1-p)+(p-0.5)**2)
```

```python
pp=np.linspace(0.5,1,100)
plt.figure(figsize=(4,4))
plt.plot(pp,pp,"k--",label="p")
plt.plot(pp,lower_bound_fun(0.01)(pp),label="$c=0.01$")
plt.plot(pp,lower_bound_fun(0.1)(pp),label="$c=0.1$")
plt.plot(pp,lower_bound_fun(0.2)(pp),label="$c=0.2$")
plt.xlabel('$p$')
plt.legend()
#plt.xlim(-0.1,1.1)
#plt.ylim(-0.1,1.1)
plt.savefig("corr.png",dpi=300,bbox_inches='tight',pad_inches=0.1)
```

```python
pp=np.linspace(0.5,1,100)
plt.figure(figsize=(4,4))
plt.plot(pp,pp,"k-",label="$p$",alpha=0.5)
plt.plot(pp,lower_bound_fun(0.01)(pp),"k-",label="$c=0.01$",alpha=0.9)
plt.plot(pp,lower_bound_fun(0.1)(pp),"k--",label="$c=0.1$",alpha=0.9)
plt.plot(pp,lower_bound_fun(0.2)(pp),"k-.",label="$c=0.2$",alpha=0.9)
plt.xlabel('$p$')
plt.legend(handlelength=3)
plt.xticks([0.5+i*0.1 for i in range(6)])
plt.yticks([0.5+i*0.1 for i in range(6)])
plt.xlim(0.45,1.05)
plt.ylim(0.45,1.05)
plt.savefig("corr.png",dpi=300,bbox_inches='tight',pad_inches=0.1)
```
